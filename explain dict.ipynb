```python
#Iterating a dictionary
#We can iterate through a dictionary using a for-loop and access the individual keys and their corresponding values
dict1={"key":"value","key1":"value1"}
#printing only keys
for key in dict1:
    print(key)
print("___________________________________")    
#for key,value in dict1:---> Error 
#    print(value)
#to print value only 
for key in dict1:
    print(dict1[key])
print("___________________________________")    
# Loop through both keys and values in the dictionary
for key, value in dict1.items():
    print(key)        # Print the current key
    print(value)      # Print the value associated with the current key
    print(key, value) # Print both the key and its value on the same line
print("___________________________________")  
for key,value in dict1.items():
    print(key,value)
print("___________________________________")  
for key,value in dict1.items():
    print(key)
    print(value)
print("___________________________________") 
print("___________________________________")  
for key, value in dict1.items():
    print(type(key))      #str
    print(type(value))    #str
    x=(key, value)
    print(type(x))        #tuple 
```

```python
#Sort dictionary
#built-in method sorted() will sort the keys in the dictionary and returns a sorted list
dict1 = {'c': 45, 'b': 95, 'a': 35}

# sorting dictionary by keys
print(sorted(dict1.items()))
# Output [('a', 35), ('b', 95), ('c', 45)]

# sort dict eys
print(sorted(dict1))
# output ['a', 'b', 'c']

# sort dictionary values
print(sorted(dict1.values()))
# output [35, 45, 95]
```

```python
#Dictionary comprehension
#output_dictionary = {key : value for key,value in iterable [if key,value condition1]}
# calculate the square of each even number from a list and store in dict
numbers = [1, 3, 5, 2, 8]
even_squares = {x: x ** 2 for x in numbers if x % 2 == 0}
print(even_squares)
print("--------------------")
telephone_book = [1178, 4020, 5786]
persons = ['Jessa', 'Emma', 'Kelly']
telephone_Directory = {key: value for key, value in zip(persons, telephone_book)}
print(telephone_Directory)
#zip(persons, telephone_book) ------>> [('Jessa', 1178), ('Emma', 4020), ('Kelly', 5786)]
#zip() refers to the two tuples.
```
