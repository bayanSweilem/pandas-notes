 What is a DataFrame?

A **DataFrame** is a 2-dimensional labeled data structure, similar to a table in Excel.  
It’s one of the most commonly used objects in data analysis using **pandas**.

```python
import pandas as pd
```

                                                             dataframe from dict

```python

data={"name":["Bayan","Layan","Lubna","Samir","Eman","Haitham"],
      "Age":[19,22,18,55,45,10],
      "Major":["Ai & data","dentist","eng.","reporter","arabic","middle school"]
     }
```

```python
dfdict=pd.DataFrame(data)
print(dfdict.shape)
dfdict
```

                                                           dataframe from list

```python
data2=[
    ["Bayan","Layan","Lubna","Samir","Eman","Haitham"],
    [19,22,18,55,45,10],
    ["Ai & data","dentist","eng.","reporter","arabic","middle school"]
]
dflist=pd.DataFrame(data2)
dflist.shape
dflist
```

```python
data3=[
    ["Bayan","Layan","Lubna","Samir","Eman","Haitham"],
    [19,22,18,55,45,10],
    ["Ai & data","dentist","eng.","reporter","arabic","middle school"]
]
df2 = pd.DataFrame(data3, columns=["P1", "P2", "P3","P4","P5","P6"])
df2
```

```python
data = [
    ["Ali", 21, "AI"],
    ["Rana", 23, "Data Science"],
    ["Tariq", 22, "Cybersecurity"],
    ["Lina", 24, "AI"]
]

df2 = pd.DataFrame(data, columns=["Name", "Age", "Major"])
df2
```

```python
print("shape",F1_DRIVERS.shape)
print("\ncol:",F1_DRIVERS.columns)
print("\ntype",F1_DRIVERS.dtypes)
```

```python
F1_DRIVERS.info()
# This calls (executes) the method.
# It actually prints the summary information about the DataFrame.
```

```python
F1_DRIVERS.info
# This refers to the method object itself — not the result of calling it.
# It just shows that there’s a function named .info, without actually running it.

# To check what the method is
# To pass the function to another function
# To check if the DataFrame has the attribute
```

Code	           What it does	               Output
F1_DRIVERS.info	   References the method	  <bound method DataFrame.info of ...>
F1_DRIVERS.info()	Executes the method	      Prints DataFrame summary

```python
print(F1_DRIVERS.values) #Returns the raw data as a NumPy array
print(F1_DRIVERS.columns)#The result is a pandas Index object, not a normal list.
# It behaves like a list in many ways, but it’s more powerful for internal pandas operations.
print(list(F1_DRIVERS.columns))#To convert it to a simple Python list
print("\nfirst col:",F1_DRIVERS.columns[0])
print("\n",F1_DRIVERS.index)

# | Code                       | What it returns   | Type           |
# | -------------------------- | ----------------- | -------------- |
# | `F1_DRIVERS.columns`       | Column names      | `pandas.Index` |
# | `list(F1_DRIVERS.columns)` | Column names      | `list`         |
# | `F1_DRIVERS.columns[0]`    | First column name | `'Name'`       |
```
